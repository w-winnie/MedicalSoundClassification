# Pre processing Audio

## Given Functions

```python
def pad_audio(y, target_length):
    """Pads the audio with zeros to match the target length."""
    if len(y) < target_length:
        padding = target_length - len(y)
        y = np.pad(y, (0, padding), 'constant')
    return y

def truncate_audio(y, target_length):
    """Truncates the audio to match the target length."""
    if len(y) > target_length:
        y = y[:target_length]
    return y

def adjust_audio(y, target_length=32000):
    """Adjusts the audio to match the target length by padding or truncating."""
    if len(y) < target_length:
        return pad_audio(y, target_length)
    else:
        return truncate_audio(y, target_length)

def ensure_sample_rate(original_sample_rate, waveform,
                       desired_sample_rate=16000):
    """Resample waveform if required."""
    if original_sample_rate != desired_sample_rate:
        desired_length = int(round(float(len(waveform)) /
                                  original_sample_rate * desired_sample_rate))
        waveform = scipy.signal.resample(waveform, desired_length)
    return desired_sample_rate, waveform
```

```python
def segment_cough_sound(signal, sr, cough_threshold=0.05, min_cough_duration=0.1, padding=0.05):

    hop_length = int(min_cough_duration*sr)
    if len(signal.shape) > 1:
        signal = np.mean(signal, axis=1)

    energy = rms(y=signal, hop_length=hop_length)[0]

    # Normalize the energy values
    normalized_energy = (energy - np.min(energy)) / (np.max(energy) - np.min(energy))

    # Set the energy threshold for event detection
    cough_threshold = np.max(normalized_energy) * cough_threshold
    min_cough_samples = round(sr * min_cough_duration)


    # Find the cough segments
    cough_segments = []
    event_start = None

    for i, value in enumerate(normalized_energy):
        if value >= cough_threshold:
            if event_start is None:
                event_start = i*hop_length
        else:
            if event_start is not None:
                cough_duration = i*hop_length - event_start
                if cough_duration >= min_cough_samples:
                    event_end = i*hop_length + int(padding * sr)
                    event_start -= int(padding * sr)
                    event_start = max(event_start, 0)
                    cough_segments.append(signal[event_start: event_end+1])
                event_start = None

    # Convert cough segments to time in seconds
    # cough_segments = [(start / sr, end / sr) for start, end in cough_segments]

    return cough_segments
```

## Extract segments

```python
import librosa
import librosa.display
import soundfile as sf

# Dictionary to store segment info
cough_segments_data = {}

# Process audio files
target_sample_rate = 16000  # Standard sample rate
target_length = 32000  # Fixed length for training

output_dir = "processed_coughs2"
os.makedirs(output_dir, exist_ok=True)
processed_files=[]

for i, audio_path in enumerate(all_audio_files):  #all_audio_files[:5])
    folder_name = os.path.basename(os.path.dirname(audio_path))  # Gets '004db89f19f9b'
    file_name1 = os.path.basename(audio_path)  # Gets 'cough.wav'
    file_name = f"{folder_name}_{file_name1}"
    y, sr = librosa.load(audio_path, sr=None)
    
    if y is None or len(y) == 0:
        print(f"Skipping empty file: {file_name}")
        continue
        
    sr, y = ensure_sample_rate(sr, y, target_sample_rate)
    
    if y is None or len(y) == 0:
        print(f"Skipping after resampling (empty data): {file_name}")
        continue
        
    y = adjust_audio(y, target_length)

    # Extract cough segments
    print(f"Processing: [{i+1}] {file_name}")
    cough_segments = segment_cough_sound(y, sr, cough_threshold=0.05, min_cough_duration=0.1)

    if not cough_segments:
        print(f"[{i+1}] {file_name} -> No cough segments detected")
        continue  # Skip if no segments found
    
    segment_times = [] 
    for j, segment in enumerate(cough_segments):
        if len(segment) == 0:
            continue
        segment_path = os.path.join(output_dir, f"{folder_name}_{i}_{j}_{file_name1}")
        sf.write(segment_path, segment, sr)
        print(f"Saved: {segment_path}")

        start_time = y.tolist().index(segment[0]) / sr
        end_time = start_time + len(segment) / sr
        segment_times.append((start_time, end_time))
    
    processed_files.append(file_name)
    cough_segments_data[file_name] = (y, sr, segment_times)

print("Processing complete")
```

```python
import json
import pickle
processed_files_path = os.path.join(output_dir, "processed_files_nonempty.json")
with open(processed_files_path, "w") as f:
    json.dump(list(processed_files), f)
print(f"Saved processed files list: {processed_files_path}")

cough_segments_data_path = os.path.join(output_dir, "cough_segments_data_nonempty.pkl")
with open(cough_segments_data_path, "wb") as f:
    pickle.dump(cough_segments_data, f)
print(f"Saved cough segment data: {cough_segments_data_path}")
```

## Inspect

```python
def plot_waveforms(file_name):
    if file_name not in cough_segments_data:
        print(f"No data found for {file_name}")
        return
    
    y, sr, segment_times = cough_segments_data[file_name]
    
    plt.figure(figsize=(12, 5))
    librosa.display.waveshow(y, sr=sr, alpha=0.6, label="Full Audio")
    
    # Highlight detected cough segments
    for start, end in segment_times:
        plt.axvspan(start, end, color='red', alpha=0.3, label="Detected Cough" if start == segment_times[0][0] else "")

    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")
    plt.title(f"Full Waveform: {file_name}")
    plt.legend()
    plt.show()

    # Plot each cough segment separately
    for i, (start, end) in enumerate(segment_times):
        segment = y[int(start * sr):int(end * sr)]  # Extract segment from full audio
        plt.figure(figsize=(8, 3))
        librosa.display.waveshow(segment, sr=sr, alpha=0.8, label=f"Cough Segment {i+1}")
        plt.xlabel("Time (seconds)")
        plt.ylabel("Amplitude")
        plt.title(f"Cough Segment {i+1} from {file_name}")
        plt.legend()
        plt.show()
```
